# Task 1

This will load the *fashion_mnist* dataset and define the training and testing data.

```python
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()

train_images = train_images / 255.0
test_images = test_images / 255.0
```

# Task 2

This will build the model and train (fit) it with training images.

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(128, activation=tf.nn.relu),
    keras.layers.Dense(10, activation=tf.nn.softmax)
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

model.fit(train_images, train_labels, epochs=3)
```

 This will find the test accuracy.

```python
test_loss, test_acc = model.evaluate(test_images, test_labels)

print('Test accuracy:', test_acc)
```

# Task 3

This will randomly select some testing images and classify them.

```python
predictions = model.predict(test_images)

num_rows = 5
num_cols = 4
num_images = num_rows*num_cols

plt.figure(figsize=(3*num_cols, 3*num_rows))

for i in range(num_images):
    plt.subplot(num_rows, num_cols, i+1)
    
    plot_image(random.randint(0, 10000), predictions, test_labels, test_images)
plt.show()
```
